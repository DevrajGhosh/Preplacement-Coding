```python
# 1. Insert an element at end of array
class Solution:
    def insertAtEnd(self, arr, val):
        arr.append(val)
        return arr

arr = list(map(int, input().split()))
val = int(input())
print(Solution().insertAtEnd(arr, val))
```

```python
# 2. Find largest element in array
class Solution:
    def largest(self, arr):
        a = arr[0]
        for i in range(1, len(arr)):
            if a < arr[i]:
                a = arr[i]
        return a

arr = list(map(int, input().split()))
print(Solution().largest(arr))
```

```python
# 3. Find second largest element
class Solution:
    def getSecondLargest(self, arr):
        n = len(arr)
        lg = float('-inf')  # largest
        sl = float('-inf')
        for i in arr:
            if(lg<i):
                sl = lg
                lg = i
            elif(sl<i and i!=lg):
                sl = i
        if(sl!=float('-inf')):
            return sl
        else:
            return -1

arr = list(map(int, input().split()))
print(Solution().getSecondLargest(arr))
```

```python
# 4. Move all zeros to end
class Solution:
    def pushZerosToEnd(self, arr):
        n = len(arr)
        count = 0

        for i in range(n):
            if arr[i] != 0:
                arr[count] = arr[i]
                count += 1

        while count < n:
            arr[count] = 0
            count += 1

        return arr

arr = list(map(int, input().split()))
print(Solution().pushZerosToEnd(arr))
```

```python
# 5. Rotate array by one
class Solution:
    def rotate(self, arr):
        n = len(arr)
        last = arr[n-1]
        for i in range(n-2, -1, -1):
            arr[i+1] = arr[i]
        arr[0] = last
        return arr

arr = list(map(int, input().split()))
print(Solution().rotate(arr))
```

```python
# 6. Check if array is sorted
class Solution:
    def isSorted(self, arr) -> bool:
        brr = sorted(arr)
        check = 1
        n = len(arr)
        for i in range(n):
            if arr[i] != brr[i]:
                check = 0
                break
        return check == 1

n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter the elements separated by space: ").split()))

sol = Solution()
print( sol.isSorted(arr))
```

```python
# 7. Reverse a string
class Solution:
    def reverseString(self, s: str) -> str:
        # code here
        brr = []
        n = len(s)
        for i in range(n-1, -1, -1):
            brr.append(s[i])
        return "".join(brr)

s = input()
print(Solution().reverseString(s))
```

```python
# 8. Check if string is palindrome
class Solution:
    def isPalindrome(self, s: str) -> bool:
        i, j = 0, len(s) - 1
        while i < j:
            if s[i] != s[j]:
                return False
            i += 1
            j -= 1
        return True

s = input()
print(Solution().isPalindrome(s))
```

```python
# 9. Count frequency of array elements (1 to n)
class Solution:
    def frequencyCount(self, arr):
        # code here
        n = len(arr)
        freq = [0] * n
        for num in arr:
            if 1 <= num <= n:
                freq[num-1] += 1
        return freq

arr = list(map(int, input().split()))
print(Solution().frequencyCount(arr))
```

```python
# 10. reverse an array
def reverseArray(arr):
    return arr[::-1]


arr = list(map(int, input().split()))

reversed_arr = reverseArray(arr)

print(' '.join(map(str, reversed_arr)))
```
